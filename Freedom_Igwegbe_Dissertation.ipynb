```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker

# Path to the folder containing the datasets
folder_path = '/content/drive/MyDrive/My datasets'

# Function to perform EDA on a dataset
def perform_eda(file_path):
    # Read the dataset
    df = pd.read_csv(file_path)

    # Display basic information
    print(f"Dataset: {os.path.basename(file_path)}")
    print("First few rows:")
    print(df.head())
    print("\nSummary statistics:")
    print(df.describe())
    print("\nMissing values:")
    print(df.isnull().sum())
    print("\n---------------------------------\n")

# Iterate through each file in the folder and perform EDA
for root, dirs, files in os.walk(folder_path):
    for file in files:
        if file.endswith('.csv'):
            file_path = os.path.join(root, file)
            perform_eda(file_path)
```

# **Predicting Energy usage with the rate of increase used to predict population increase**
